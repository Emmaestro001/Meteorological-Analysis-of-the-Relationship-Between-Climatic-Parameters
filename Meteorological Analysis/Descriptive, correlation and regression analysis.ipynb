```python
#IMPORT NECESSARY PACKAGES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#IMPORT ANNUAL DATA FOR EACH VARIABLE
result = pd.read_excel("EPT.xlsx", 'Sheet2')
```

```python
#VIEW DATA
result.head()
```

```python
#CONDUCT DESCRIPTIVE ANALYSIS OF ANNUAL DATA
result.describe(include='all')
```

```python
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
DESCRIBE=result.describe(include='all')
```

```python
#SAVE DESCRIPTIVE ANALYSIS DATA
DESCRIBE.to_excel('ANNUAL DESCRIBE.xlsx')
```

```python
#CONDUCT CORRELATION ANALYSIS ON ANNUAL DATA

corr = result.corr()
corr
```

```python
#SAVE CORRELATION MATRIX TABLE TO EXCEL
corr.to_excel('ANNUAL CORR.xlsx')
```

```python
#VISUALIZE ANNUAL CORRELATION MATRIX TABLE 
fig = plt.figure(figsize=(8,8)) #set figure size
plt.matshow(corr, cmap='RdBu', fignum=fig.number) #plot matrix
plt.xticks(range(len(corr.columns)), corr.columns, rotation='vertical');
plt.yticks(range(len(corr.columns)), corr.columns);

cb = plt.colorbar() #include colorbar
cb.ax.tick_params(labelsize=10) #tick label size
```
