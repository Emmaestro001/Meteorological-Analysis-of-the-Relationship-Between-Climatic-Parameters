```python
#IMPORT NECESSARY PYTHON PACKAGES    

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Note that the excel sheets for each season that will be imported in the suceeding cells have been edited in Excel such that the time/month that is not within the season has no number, i.e., NaN. This can be seen from the excel workbook 'NEW_EPT.xlsx' 

```python
#IMPORT EXCEL SHEET FOR DRY SEASON
dry = pd.read_excel("NEW_EPT.xlsx", 'DRY', index_col='DATE')
dry.head()
```

```python
#IMPORT EXCEL SHEET FOR WET SEASON
wet = pd.read_excel("NEW_EPT.xlsx", 'WET', index_col='DATE')
wet.head()
```

BEGIN PLOTTING WET AND DRY SEASON TRENDS FOR EACH PARAMETER IN STUDY   

```python
plt.plot(dry['Atm. Pressure'], 'k.-', label="Dry season")
plt.plot(wet['Atm. Pressure'], 'r.:', label="Wet season")


plt.xlabel('Time (years)')
plt.ylabel("Atmospheric Pressure (hPa)")
plt.legend(bbox_to_anchor=(0.5, 1), loc="upper right")
plt.show()

plt.savefig('Atm. Pressure.png', dpi=300)
```
